## Bivariate EDA

```python
ct = pd.crosstab(df['Warehouse_block'], df['Reached.on.Time_Y.N'])

ct.plot(kind='bar')
plt.xlabel('Warehouse_block')
plt.ylabel('Count')
plt.title('Warehouse_block vs Reached on Time')
plt.legend(title='Reached on Time')
plt.show()
```

```python
df.head()
```

```python
ct = pd.crosstab(df['Mode_of_Shipment'], df['Reached.on.Time_Y.N'])

ct.plot(kind='bar')
plt.xlabel('Mode_of_Shipment')
plt.ylabel('Count')
plt.title('Mode_of_Shipment vs Reached on Time')
plt.legend(['Not On Time', 'On Time'])
plt.show()
```

```python
sns.countplot(
    data=df,
    x='Customer_care_calls',
    hue='Reached.on.Time_Y.N'
)
plt.xlabel('Customer Care Calls')
plt.ylabel('Count')
plt.title('Customer Care Calls vs Reached on Time')
plt.show()
```

```python
sns.countplot(
    data=df,
    x='Customer_rating',
    hue='Reached.on.Time_Y.N'
)
plt.xlabel('Customer_rating')
plt.ylabel('Count')
plt.title('Customer_rating vs Reached on Time')
plt.show()
```

```python
plt.figure()
df.boxplot(
    column='Cost_of_the_Product',
    by='Reached.on.Time_Y.N'
)
plt.xlabel('Reached on Time (0 = No, 1 = Yes)')
plt.ylabel('Cost_of_the_Product')
plt.title('Cost_of_the_Product vs Reached on Time')
plt.suptitle('')
plt.show()
```

```python
sns.countplot(
    data=df,
    x='Prior_purchases',
    hue='Reached.on.Time_Y.N'
)
plt.xlabel('Prior_purchases')
plt.ylabel('Count')
plt.title('Prior_purchases vs Reached on Time')
plt.show()
```

```python
sns.countplot(
    data=df,
    x='Prior_purchases',
    hue='Reached.on.Time_Y.N'
)
plt.xlabel('Prior_purchases')
plt.ylabel('Count')
plt.title('Prior_purchases vs Reached on Time')
plt.show()
```

```python
ct = pd.crosstab(df['Product_importance'], df['Reached.on.Time_Y.N'])

ct.plot(kind='bar')
plt.xlabel('Product_importance')
plt.ylabel('Count')
plt.title('Product_importance vs Reached on Time')
plt.legend('Not on Time','Reached on Time')
plt.show()
```

```python
ct = pd.crosstab(df['Gender'], df['Reached.on.Time_Y.N'])

ct.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender vs Reached on Time')
plt.legend('Not on Time','Reached on Time')
plt.show()
```

```python
plt.figure()
df.boxplot(
    column='Discount_offered',
    by='Reached.on.Time_Y.N'
)
plt.xlabel('Reached on Time (0 = No, 1 = Yes)')
plt.ylabel('Discount_offered')
plt.title('Discount_offered vs Reached on Time')
plt.suptitle('')
plt.show()
```

## Multivariate EDA

```python
g = sns.catplot(
    data=df,
    x='Warehouse_block',
    hue='Reached.on.Time_Y.N',
    col='Mode_of_Shipment',
    kind='count',
    height=4,
    aspect=1
)

g.set_axis_labels('Warehouse Block', 'Count')
g.fig.suptitle('Warehouse Block vs Reached on Time by Mode of Shipment', y=1.05)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='Weight_in_gms',
    y='Discount_offered',
    hue='Reached.on.Time_Y.N',
    alpha=0.5
)
plt.xlabel('Weight in gms')
plt.ylabel('Discount Offered')
plt.title('Weight vs Discount by Delivery Status')
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x='Reached.on.Time_Y.N',
    y='Weight_in_gms',
    hue='Product_importance'
)
plt.xlabel('Reached on Time (0 = No, 1 = Yes)')
plt.ylabel('Weight in gms')
plt.title('Weight vs Reached on Time by Product Importance')
plt.show()
```

```python
sns.catplot(
    data=df,
    x='Product_importance',
    hue='Reached.on.Time_Y.N',
    col='Mode_of_Shipment',
    kind='count',
    height=4,
    aspect=1
)
plt.show()
```

```python
sns.catplot(
    data=df,
    x='Product_importance',
    hue='Reached.on.Time_Y.N',
    col='Gender',
    kind='count',
    height=4,
    aspect=1
)
plt.show()
```

## Heat Map

```python
plt.figure(figsize=(12,8))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.show()
```
